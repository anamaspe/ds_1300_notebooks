```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../data/state_facts.tsv', delimiter='\t')
data.head()
```

```python
plot = data[['USPS_code','Area_tot','Area_land','Area_water']].plot(kind='bar',figsize=(10,10), colormap="cool") 
plot.set_xticklabels(data['USPS_code'])
plot.set_xlabel('State/Territory Abbreviation')
plot.set_ylabel('Area')
plot.set_title('Land, Water, and Total Area of US States and Territories')
plot.legend(['Total Area','Land Area','Water Area']);
```

My bar graph shows the land, water and total areas of the US States and territories.

I decided to use a bar graph to clarly display the different areas (land, water, and total) of the US states and territories, and I chose to use state abbreviations on the bottom instead of the full state names to save space. Lastly, I chose the color "cool" from matplotlib and renamed the values in the legend to make the graph more organized and clean.
